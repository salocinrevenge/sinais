```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-3, 3, 1000)

kmax = 1000
y = np.zeros(len(x))

T = 7
coeficientes = []
tau1 = 4
for k in range(-kmax,kmax):
    if k == 0:
        coeficiente = tau1/T
    else:
        coeficiente = (np.sin(k*np.pi*tau1/T)/(np.pi*k))
    # coeficiente = T * (coeficiente **2) * -2
    coeficientes.append(coeficiente)

for k in range(-kmax,kmax):
    parcela = coeficientes[k+kmax]*np.exp(1j * k*x*np.pi*2/T)
    y = y + parcela


# Cria o gráfico
plt.figure(figsize=(8, 4))  # Define o tamanho da figura
plt.plot(x, y, label='Sinal')  # Plota a senoide
plt.xlabel('t')  # Rótulo do eixo x
plt.ylabel('Amplitude')  # Rótulo do eixo y
plt.grid(True)  # Ativa a grade no gráfico
plt.legend()  # Mostra a legenda

# Exibe o gráfico
plt.show()



# Cria o gráfico dos coeficientes
plt.figure(figsize=(8, 4))  # Define o tamanho da figura
plt.plot([i for i in range(-20,20)], coeficientes[kmax-20:kmax+20], label='Coeficientes', color = "tab:red")  # Plota a senoide
plt.xlabel('k')  # Rótulo do eixo x
plt.ylabel('Amplitude')  # Rótulo do eixo y
plt.grid(True)  # Ativa a grade no gráfico
plt.legend()  # Mostra a legenda

# Exibe o gráfico
plt.show()
```

```python
def curva(N = 10000, cor = "tab:blue"):
    import numpy as np
    import matplotlib.pyplot as plt


    x = np.linspace(-10, 10, 1000)

    kmax = N
    y = np.zeros(len(x))

    T = 7
    coeficientes1 = []
    tau1 = 4
    for k in range(-kmax,kmax+1):
        if k == 0:
            coeficiente = tau1/T
        else:
            coeficiente = (np.sin(k*np.pi*tau1/T)/(np.pi*k))
        coeficientes1.append(coeficiente)

    coeficientes2 = []
    tau2 = 2
    for k in range(-kmax,kmax+1):
        if k == 0:
            coeficiente = tau2/T
        else:
            coeficiente = (np.sin(k*np.pi*tau2/T)/(np.pi*k))
        coeficientes2.append(coeficiente)

    coeficientes3 = []
    tau3 = 1
    for k in range(-kmax,kmax+1):
        if k == 0:
            coeficiente = tau3/T
        else:
            coeficiente = (np.sin(k*np.pi*tau3/T)/(np.pi*k))
            # coeficiente = 2 * np.exp(1j*k*np.pi/2) * (np.sin(k*np.pi/2)/(1j*(np.pi*k)**2)) # triangular
        coeficiente = T * (coeficiente **2) * -2
        coeficientes3.append(coeficiente)

    coeficientes = []
    for k in range(-kmax,kmax+1):
        coeficiente = coeficientes1[k+kmax]+coeficientes2[k+kmax]+coeficientes3[k+kmax]
        coeficientes.append(coeficiente)
        parcela = coeficiente*np.exp(1j * k*x*np.pi*2/T)
        y = y + parcela

    xOficial = [-10,-9,-9,-8,-8,-7,-6,-6,-5,-5,-4,-3,-2,-2,-1,-1,0,1,1,2,2,3,4,5,5,6,6,7,8,8,9,9,10]
    yOficial = [0,   0, 1, 1, 2, 0, 2, 1, 1, 0, 0, 0, 0, 1, 1, 2,0,2,1,1,0,0,0,0,1,1,2,0,2,1,1,0,0]

    # Cria o gráfico
    plt.figure(figsize=(8, 4))  # Define o tamanho da figura
    plt.plot(xOficial, yOficial, label='Sinal x', color = "tab:blue")  # Plota a senoide
    plt.plot(x, y, label=f'Série de fourier para N = {N}', color = cor)  # Plota a senoide
    plt.xlabel('t')  # Rótulo do eixo x
    plt.ylabel('Amplitude')  # Rótulo do eixo y
    plt.grid(True)  # Ativa a grade no gráfico
    plt.legend()  # Mostra a legenda

    # Exibe o gráfico
    plt.show()

    # cria a funcao y = -2x
    linear = -2*x

    erro = 0
    # erro 1 é o erro da curva y em relacao a curva zeros de -3.5 a -2
    for i in range(len(x)):
        if x[i] >= -3.5 and x[i] <= -2:
            erro += (y[i])**2
            continue
        if x[i] >= -2 and x[i] <= -1:
            erro += (y[i]-1)**2
            continue
        if x[i] >= -1 and x[i] <= 0:
            erro += (y[i]-linear[i])**2
            continue
    erro = erro*2/7
    print(f"Erro: {erro}")
```
